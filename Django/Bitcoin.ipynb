# Are there too many competitors in the market that challenge bitcoin?
Bitcoin may be the first, but it is not the only crypto currency in the market. The graph below shows that Bitcoin is still leading, but crytpocurrencies like Ethereum or Litecoin are slowly catching up to Bitcoin's market price. But just how fast is it approaching?

```python
ax = df1.groupby(['name'])['market'].last().sort_values(ascending=False).head(10).sort_values().plot(kind='bar')
ax.set_xlabel("Market cap")
plt.title("Top 10 Currencies by Market Cap")
plt.show()
```

```python
dfEFull = df1.loc[df1['name'] == 'Ethereum']
dfE = dfEFull['market']
dfE.plot(color='purple')
plt.xticks([])
plt.ylabel("Price")
plt.xlabel("Time")
plt.title("Ethereum Market Prices")
plt.show()
```

#### Ethereum is actually on a downward trend as well in terms of market price

```python
dfRFull = df1.loc[df1['name'] == 'Ripple']
dfR = dfRFull['market']
dfR.plot()
plt.xticks([])
plt.ylabel("Price")
plt.xlabel("Time")
plt.title("Ripple Market Prices")
plt.show()
```

#### Even Ripple isn't safe from this market crash

# What does this mean?

# What about real world currency?
Let's see how bitcoin then compares to real world currency and such. Look at the US Dollar.

```python
dfUSAFull = pd.read_csv('DTWEXB.csv', na_values='.')
dfUSA = dfUSAFull['VALUE']
dfUSA.plot()
plt.xticks([])
plt.ylabel("Price")
plt.xlabel("Time")
plt.title("Trade Weighted U.S. Dollar Index")
plt.show()
```

### Looks like the US dollar is on a downward trend as well, but it's drop isn't as drastic as Bitcoin's. This stability is the reason that people prefer using traditional currencies over Bitcoin.
In a book by Ricky W. Griffin called 'Organizational Behavior: Managing People and Organizations', people have a trait called 'Uncertainty Avoidance', also known as a 'Preference for Stability,' which is the extent that people will feel threatened by the unknown and prefer to in clear and unambiguous situations. People continue to invest in the dollar because it is seen as a well known, stable, currency.
